#QUESTION 1

```python
import pandas as pd

# Load dataset
df = pd.read_csv("dataset.csv", low_memory=False)

# Drop index-like column if present
df = df.drop(columns=["Unnamed: 0"], errors="ignore")

# Q1a: Split metadata & features
META_COLS = ["track_id", "artists", "album_name", "track_name", "track_genre"]

meta_df = df[META_COLS].copy()
X_df = df.drop(columns=META_COLS).copy()

# Q1b: One-Hot Encode Genres
genre_dummies = pd.get_dummies(meta_df["track_genre"], prefix="genre").astype("int8")
X_df = X_df.join(genre_dummies)


# Fix explicit column
if "explicit" in X_df.columns:
    X_df["explicit"] = (
        pd.to_numeric(X_df["explicit"], errors="coerce")
        .fillna(0)
        .astype("int64")
    )

# Final sanity check
non_numeric = X_df.select_dtypes(exclude="number").columns.tolist()
print("Nonnumeric columns left:", non_numeric)
print("meta_df shape:", meta_df.shape)
print("X_df shape:", X_df.shape)
```

#QUESTION 2

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_df)

print("X_scaled shape:", X_scaled.shape)
print("Mean (first 5 features):", X_scaled.mean(axis=0)[:5])
print("Std  (first 5 features):", X_scaled.std(axis=0)[:5])
```

#QUESTION 3

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Q3: K-Means + SSE
k_values = [2, 3, 5, 7, 10, 15, 20]
sse = []

for k in k_values:
    kmeans = KMeans(
        n_clusters=k,
        init="k-means++",
        n_init=10,
        random_state=42
    )
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure()
plt.plot(k_values, sse, marker="o")
plt.xlabel("Number of clusters (k)")
plt.ylabel("SSE (Inertia)")
plt.title("Elbow Method for K-Means")
plt.grid(True)
plt.show()

for k, val in zip(k_values, sse):
    print(f"k={k}, SSE={val:.2f}")
```

#QUESTION 4

```python
from sklearn.cluster import KMeans

# Q4: Final K-Means
BEST_K = 10

kmeans_final = KMeans(
    n_clusters=BEST_K,
    init="k-means++",
    n_init=10,
    random_state=42
)

cluster_labels = kmeans_final.fit_predict(X_scaled)

# attach labels
X_df["cluster"] = cluster_labels
meta_df["cluster"] = cluster_labels

print("Number of clusters:", len(set(cluster_labels)))
meta_df.head()
```

#QUESTION 5

```python
import matplotlib.pyplot as plt

cluster_genre_df = meta_df[["cluster", "track_genre"]]

# counts genres
genre_counts = (
    cluster_genre_df
    .groupby(["cluster", "track_genre"])
    .size()
    .reset_index(name="count")
)

cluster_sizes = (
    cluster_genre_df
    .groupby("cluster")
    .size()
    .reset_index(name="total")
)

merged = genre_counts.merge(cluster_sizes, on="cluster")

merged["percentage"] = merged["count"] / merged["total"] * 100

dominant = (
    merged
    .sort_values(["cluster", "percentage"], ascending=[True, False])
    .groupby("cluster")
    .first()
    .reset_index()
)

print(dominant[["cluster", "track_genre", "percentage"]])

plt.figure(figsize=(10, 5))
plt.bar(dominant["cluster"], dominant["percentage"])
plt.xlabel("Cluster")
plt.ylabel("Dominant Genre Percentage (%)")
plt.title("Dominant Genre Percentage per Cluster")
plt.xticks(dominant["cluster"])
plt.show()
```

#QUESTION 6

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

SONG_INDEX = 2

selected_meta = meta_df.iloc[SONG_INDEX]
selected_cluster = selected_meta["cluster"]

print("Selected song:")
print("Title :", selected_meta["track_name"])
print("Artist:", selected_meta["artists"])
print("Cluster:", selected_cluster)

cluster_indices = meta_df.index[meta_df["cluster"] == selected_cluster].tolist()

X_cluster = X_scaled[cluster_indices]
X_selected = X_scaled[SONG_INDEX].reshape(1, -1)

distances = euclidean_distances(X_selected, X_cluster)[0]

dist_df = meta_df.loc[cluster_indices].copy()
dist_df["distance"] = distances

dist_df = dist_df[dist_df.index != SONG_INDEX]

top3 = dist_df.nsmallest(3, "distance")

print("\nTop-3 similar songs (Euclidean distance):")
print(top3[["track_name", "artists", "distance"]])
```

#QUESTION 7

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_distances

SONG_INDEX = 2

selected_meta = meta_df.iloc[SONG_INDEX]
selected_cluster = selected_meta["cluster"]

print("Selected song:")
print("Title :", selected_meta["track_name"])
print("Artist:", selected_meta["artists"])
print("Cluster:", selected_cluster)

cluster_indices = meta_df.index[meta_df["cluster"] == selected_cluster].tolist()

X_cluster = X_scaled[cluster_indices]
X_selected = X_scaled[SONG_INDEX].reshape(1, -1)

distances = cosine_distances(X_selected, X_cluster)[0]

dist_df = meta_df.loc[cluster_indices].copy()
dist_df["distance"] = distances
dist_df = dist_df[dist_df.index != SONG_INDEX]

top3 = dist_df.nsmallest(3, "distance")

print("\nTop-3 similar songs (Cosine distance):")
print(top3[["track_name", "artists", "distance"]])
```

#QUESTION 8

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances, cosine_distances

NEW_K = 15
SONG_INDEX = 0

kmeans_alt = KMeans(n_clusters=NEW_K, init="k-means++", n_init=10, random_state=42)
new_labels = kmeans_alt.fit_predict(X_scaled)

X_df["cluster_alt"] = new_labels
meta_df["cluster_alt"] = new_labels

cluster_genre_df = meta_df[["cluster_alt", "track_genre"]]

genre_counts = cluster_genre_df.groupby(["cluster_alt", "track_genre"]).size().reset_index(name="count")
cluster_sizes = cluster_genre_df.groupby("cluster_alt").size().reset_index(name="total")

merged = genre_counts.merge(cluster_sizes, on="cluster_alt")
merged["percentage"] = merged["count"] / merged["total"] * 100

dominant = (
    merged
    .sort_values(["cluster_alt", "percentage"], ascending=[True, False])
    .groupby("cluster_alt")
    .first()
    .reset_index()
)

plt.figure(figsize=(10,5))
plt.bar(dominant["cluster_alt"], dominant["percentage"])
plt.xlabel("Cluster")
plt.ylabel("Dominant Genre Percentage (%)")
plt.title("Dominant Genre per Cluster (Alternative k)")
plt.show()

selected_meta = meta_df.iloc[SONG_INDEX]
selected_cluster = selected_meta["cluster_alt"]

cluster_indices = meta_df.index[meta_df["cluster_alt"] == selected_cluster].tolist()

X_cluster = X_scaled[cluster_indices]
X_selected = X_scaled[SONG_INDEX].reshape(1, -1)

euc_dist = euclidean_distances(X_selected, X_cluster)[0]
cos_dist = cosine_distances(X_selected, X_cluster)[0]

euc_df = meta_df.loc[cluster_indices].copy()
euc_df["distance"] = euc_dist
euc_df = euc_df[euc_df.index != SONG_INDEX]
top3_euc = euc_df.nsmallest(3, "distance")

cos_df = meta_df.loc[cluster_indices].copy()
cos_df["distance"] = cos_dist
cos_df = cos_df[cos_df.index != SONG_INDEX]
top3_cos = cos_df.nsmallest(3, "distance")

print("Selected song:")
print(selected_meta[["track_name", "artists", "cluster_alt"]])
print("\nTop-3 Euclidean:")
print(top3_euc[["track_name", "artists", "distance"]])
print("\nTop-3 Cosine:")
print(top3_cos[["track_name", "artists", "distance"]])
```

#QUESTION 9

```python
from sklearn.cluster import DBSCAN
import numpy as np

dbscan = DBSCAN(eps=3.0, min_samples=10, metric="euclidean")
db_labels = dbscan.fit_predict(X_scaled)

meta_df["cluster_dbscan"] = db_labels
X_df["cluster_dbscan"] = db_labels

n_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)
n_noise = list(db_labels).count(-1)

print("DBSCAN clusters (excluding noise):", n_clusters)
print("Noise points:", n_noise)

print(meta_df["cluster_dbscan"].value_counts().sort_index())
```
